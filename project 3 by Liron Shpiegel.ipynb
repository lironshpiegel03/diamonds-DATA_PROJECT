t1 - carat & price statistics by daimond cut 

```python
#this means that all the rows  are grouped together based on the unique values in the 'cut' column.
diamond_stats = diamonds_df.groupby('cut').agg(
{
    'carat': ['mean', 'median', 'max', 'min'],
    'price': ['mean', 'median', 'max', 'min']
}
#the agg function is used to apply multiple aggregation functions to each group Specifically :'carat', 'price' columns 
).reset_index()
# Flatten the MultiIndex columns
diamond_stats.columns = ['cut', 'mean_Carat', 'median_Carat', 'max _carat',
                          'min_carat', 'mean_price', 'median_price', 'max _price', 'min_price']
print(diamond_stats)
```

t2 - carat & price statistics by daimond clarity 

```python
#this means that all the rows  are grouped together based on the unique values in the 'clarity' column.
clarity_stats = diamonds_df.groupby('clarity').agg(
{
    'carat': ['mean', 'median', 'max', 'min'],
    'price': ['mean', 'median', 'max', 'min']
}
#the agg function is used to apply multiple aggregation (For example: average, median, etc...) functions to each group Specifically :'carat', 'price' columns
).reset_index()
# Flatten the MultiIndex columns
clarity_stats.columns = ['clarity', 'mean_carat', 'median_carat', 'max_carat'
                         , 'min_carat', 'mean_price', 'median_price', 'max_price', 'Min_price']
print(clarity_stats)
```

t3 - carat & price statistics by daimond color 

```python
#this means that all the rows  are grouped together based on the unique values in the 'cut' column.
diamond_stats = diamonds_df.groupby('color').agg(
{
    'carat': ['mean', 'median', 'max', 'min'],
    'price': ['mean', 'median', 'max', 'min']
}
#the agg function is used to apply multiple aggregation functions to each group Specifically :'carat', 'price' columns 
).reset_index()
# Flatten the MultiIndex columns
diamond_stats.columns = ['color', 'mean_Carat', 'median_Carat', 'max _carat',
                          'min_carat', 'mean_price', 'median_price', 'max _price', 'min_price']
print(diamond_stats)
```

t4 -count of diamonds by  color  & cut 

```python
# Create a summary table with cumulative sums and totals
summary_table = []

for color in diamonds_df['color'].unique():
    color_data = diamonds_df[diamonds_df['color'] == color]
    total_count = color_data.shape[0]
    
    # Cumulative sum for each cut within the color
    cut_counts = color_data['cut'].value_counts().sort_index()
    cumulative_sum = cut_counts.cumsum()
    
    # Append the data to the summary table
    for cut, count in cut_counts.items():
        summary_table.append([color, cut, count, cumulative_sum[cut]])
    
    # Append the total for the color
    summary_table.append([color, 'Total', total_count, total_count])

# Convert the summary table to a DataFrame
cumulative_summary_df = pd.DataFrame(summary_table, columns=['color', 'cut', 'count', 'cumulative_sum'])

# Display the table
print(cumulative_summary_df)
```

t5 the rarest & commmon clarity & color  

```python
# Define functions for most common and rarest values
def most_common(series):
    return series.mode()[0]

def rarest(series):
    return series.value_counts().idxmin()

# Calculate the most common and rarest color and clarity for each type of diamond cut
diamond_cut = diamonds_df.groupby('cut').agg(
    Most_Common_Color=('color', most_common),
    Most_Common_Clarity=('clarity', most_common),
    Rarest_Color=('color', rarest),
    Rarest_Clarity=('clarity', rarest)
).reset_index()
print(diamond_cut)
```

q1- the most expensive  damond 

```python
# Ensure that the prices are numeric
diamonds_df['price'] = pd.to_numeric(diamonds_df['price'], errors='coerce')
# selects the top 5 diamonds with the highest prices
top5_expensive = diamonds_df.nlargest(5, 'price')
# sorts the top 5 most expensive diamonds by price 
sorted_top5_expensive = top5_expensive.sort_values(by='price')
sorted_prices = sorted_top5_expensive['price'].values

plt.figure(figsize=(8, 6))
#creates a line plot connecting the points (prices) with various attributes such as color, line style, and marker size.
plt.plot(range(1, 6), sorted_prices, marker='o', linestyle='-', color='blue', linewidth=2, markersize=8)

# creates a loop that iterates over each price and adds annotations to each point.
for i, price in enumerate(sorted_prices, 1):
    plt.scatter(i, price, color='blue', zorder=5)  
    plt.text(i, price, f'${price}', fontsize=10, ha='center', va='bottom', color='black')

#sores the highest price in the variable max_price.
max_price = sorted_prices[-1]
#adds a horizontal dashed red line at the height of the highest price.
plt.axhline(y=max_price, color='red', linestyle='--', linewidth=1)
#adds text explaining that this is the highest price.
plt.text(5.5, max_price, f'Highest Price: ${max_price}', fontsize=12, color='red', va='center')

plt.xlabel('Diamond Rank', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.title('Price Increase of the Top 5 Most Expensive Diamonds', fontsize=14)
# sets the x-axis tick labels to "Diamond 1", "Diamond 2", etc.
plt.xticks(range(1, 6), ['Diamond {}'.format(i) for i in range(1, 6)], fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

**Conclusions from the graph:**
  The graph shows the price increase of the five most expensive diamonds. 
  A significant increase can be seen between diamond number 3 and diamond number 4, indicating a large price gap among the most expensive diamonds. 
  The highest price is $18,823, marked by a dashed red line. 
  This shows that there are extremely high-end diamonds on the market with extremely high prices.

q3-count of diamonds by cut 

```python
# counts the number of times each cut type appears in the 'cut' column of the DataFrame.
#resets the index, moving the data into a new table.
cut_counts = diamonds_df['cut'].value_counts().reset_index()
cut_counts.columns = ['cut', 'count']

#Scraping pattern size  
plt.figure(figsize=(10, 6))
#data comes from 'cut_counts', on the x-axis the cut types appear (cut), and on the y-axis the number (count) appears. The 'Paired' palette is used to set colors. 
barplot = sns.barplot(data=cut_counts, x='cut', y='count', palette='Paired')
#  naming the axes and the graph
plt.title('Count of Diamonds by Cut Type')
plt.xlabel('Cut Type')
plt.ylabel('Count')
# Adding the legend
for patch in barplot.patches:
    barplot.annotate(format(patch.get_height(), '.0f'), 
                     (patch.get_x() + patch.get_width() / 2., patch.get_height()), 
                     ha = 'center', va = 'center', 
                     xytext = (0, 9), 
                     textcoords = 'offset points')

plt.show()
```

**Conclusions from the graph:**
  The graph shows the amount of diamonds according to cut types.
  The "Ideal" cut is the most common with 21,551 diamonds.
  After that, the "Premium" cut with 13,791 diamonds and the "Very Good" cut with 12,082 diamonds.
  The "Fair" cut is the least common with only 1,610 diamonds.
